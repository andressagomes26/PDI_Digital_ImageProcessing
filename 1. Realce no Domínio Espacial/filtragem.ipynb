```python
# Importação das bibliotecas
import numpy as np 
import cv2
from matplotlib import pyplot as plt
import os
import glob
import pathlib
```

1. Construa os sete filtros passa-baixa definidos	abaixo e discuta sobre as mudanças de aplicações,	dimensões dos filtros e efeitos em tipos de ruídos.

```python
def load_data(datadir):
    img_array = []
    path = os.path.join(datadir)
    path_list  = list(os.listdir(path))
    for img in path_list:
        img_array.append(cv2.imread(os.path.join(path, img)))
    return img_array
        
def kernel(dim, N):
    return np.ones((dim, dim), np.float32)/N

def filterImage(img):
    img_blur_3 = cv2.filter2D(img, -1, kernel(3,9))
    img_blur_5 = cv2.filter2D(img, -1, kernel(5,25))
    img_blur_7 = cv2.filter2D(img, -1, kernel(7,49))
    return img_blur_3, img_blur_5, img_blur_7

def plotImage():
    img = load_data('imagens-Sal-e-Pimenta')
    for i in range(len(img)):
        img_blur_3, img_blur_5, img_blur_7 = filterImage(img[i])
        plt.figure(figsize=(15,12))

        plt.subplot(1, 4, 1), plt.imshow(img[i]), plt.title("Original")
        plt.xticks([]), plt.yticks([])
        
        plt.subplot(1, 4, 2), plt.imshow(img_blur_3), plt.title("Filtro da media 3x3 (N=9)")
        plt.xticks([]), plt.yticks([])

        plt.subplot(1, 4, 3),plt.imshow(img_blur_5), plt.title("Filtro da media 5x5 (N=25)")
        plt.xticks([]), plt.yticks([])

        plt.subplot(1, 4, 4),plt.imshow(img_blur_7), plt.title("Filtro da media 7x7 (N=49)")
        plt.xticks([]), plt.yticks([])
        plt.show()

plotImage()
```

### Mediana

> Filtro da	Mediana: 3x3, 5x5, 7x7

```python
def filterImageMedian(img):
    img_Medianblur_3 = cv2.medianBlur(img, 3)
    img_Medianblur_5 = cv2.medianBlur(img, 5)
    img_Medianblur_7 = cv2.medianBlur(img, 7)
    return img_Medianblur_3, img_Medianblur_5, img_Medianblur_7

def plotImageMedian():
    img = load_data('imagens-Sal-e-Pimenta')
    for i in range(len(img)):
        img_Medianblur_3, img_Medianblur_5, img_Medianblur_7 = filterImageMedian(img[i])
        plt.figure(figsize=(15,12))

        plt.subplot(1, 4, 1), plt.imshow(img[i]), plt.title("Original")
        plt.xticks([]), plt.yticks([])
        
        plt.subplot(1, 4, 2), plt.imshow(img_Medianblur_3), plt.title("Filtro da mediana 3x3")
        plt.xticks([]), plt.yticks([])

        plt.subplot(1, 4, 3),plt.imshow(img_Medianblur_5), plt.title("Filtro da mediana 5x5")
        plt.xticks([]), plt.yticks([])

        plt.subplot(1, 4, 4),plt.imshow(img_Medianblur_7), plt.title("Filtro da mediana 7x7")
        plt.xticks([]), plt.yticks([])
        plt.show()

plotImageMedian()
```
