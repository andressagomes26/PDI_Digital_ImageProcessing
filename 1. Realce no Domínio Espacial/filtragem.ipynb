> Processamento Digital de Imagens

```python
# Importação das bibliotecas
import numpy as np 
import cv2
from matplotlib import pyplot as plt
import os
import glob
import pathlib
```

```python
def load_data(datadir):
    img_array = []
    path = os.path.join(datadir)
    path_list  = list(os.listdir(path))
    for img in path_list:
        img_array.append(cv2.imread(os.path.join(path, img)))
    return img_array
```

### Filtros passa-baixa

> Definindo os kernels

```python
def kernel(dim, N):
    return np.ones((dim, dim), np.float32)/N

def kernel2(dim, N):
    return np.array([[1,2,1], [2, 4, 2], [1, 2, 1]])/N
```

### Filtro da média

> (a) Dimensões: 3x3 (N=9), 5x5 (N=25) e 7x7 (N=49)

> (b) Dimensões: 3x3 (1/16)

```python
def filterImage(img):
    img_blur_3 = cv2.filter2D(img, -1, kernel(3,9))
    img_blur_5 = cv2.filter2D(img, -1, kernel(5,25))
    img_blur_7 = cv2.filter2D(img, -1, kernel(7,49))
    img_blur_3_16 = cv2.filter2D(img, -1, kernel2(3,16))
    return img_blur_3, img_blur_5, img_blur_7, img_blur_3_16
```

> Plot imagens + filtro média 

```python
def plotImage():
    img = load_data('imagens-Sal-e-Pimenta')
    for i in range(len(img)):
        img_blur_3, img_blur_5, img_blur_7, img_blur_3_16 = filterImage(img[i])
        plt.figure(figsize=(15,12))

        plt.subplot(1, 5, 1), plt.imshow(img[i]), plt.title("Original")
        plt.xticks([]), plt.yticks([])
        
        plt.subplot(1, 5, 2), plt.imshow(img_blur_3), plt.title("Filtro da media 3x3 (N=9)")
        plt.xticks([]), plt.yticks([])

        plt.subplot(1, 5, 3),plt.imshow(img_blur_5), plt.title("Filtro da media 5x5 (N=25)")
        plt.xticks([]), plt.yticks([])

        plt.subplot(1, 5, 4),plt.imshow(img_blur_7), plt.title("Filtro da media 7x7 (N=49)")
        plt.xticks([]), plt.yticks([])
        
        plt.subplot(1, 5, 5),plt.imshow(img_blur_3_16), plt.title("Filtro da media 3x3 (1/16)")
        plt.xticks([]), plt.yticks([])
        plt.show()

plotImage()
```

### Mediana

> Filtro da	Mediana: 3x3, 5x5, 7x7

### Filtro da Mediana

> (c) Dimensões: 3x3 (N=9), 5x5 (N=25) e 7x7 (N=49)

> (b) Dimensões: 3x3 (1/16)

```python
def filterImageMedian(img):
    img_Medianblur_3 = cv2.medianBlur(img, 3)
    img_Medianblur_5 = cv2.medianBlur(img, 5)
    img_Medianblur_7 = cv2.medianBlur(img, 7)
    return img_Medianblur_3, img_Medianblur_5, img_Medianblur_7

def plotImageMedian():
    img = load_data('imagens-Sal-e-Pimenta')
    for i in range(len(img)):
        img_Medianblur_3, img_Medianblur_5, img_Medianblur_7 = filterImageMedian(img[i])
        plt.figure(figsize=(15,12))

        plt.subplot(1, 4, 1), plt.imshow(img[i]), plt.title("Original")
        plt.xticks([]), plt.yticks([])
        
        plt.subplot(1, 4, 2), plt.imshow(img_Medianblur_3), plt.title("Filtro da mediana 3x3")
        plt.xticks([]), plt.yticks([])

        plt.subplot(1, 4, 3),plt.imshow(img_Medianblur_5), plt.title("Filtro da mediana 5x5")
        plt.xticks([]), plt.yticks([])

        plt.subplot(1, 4, 4),plt.imshow(img_Medianblur_7), plt.title("Filtro da mediana 7x7")
        plt.xticks([]), plt.yticks([])
        plt.show()

plotImageMedian()
```
